```python
import pandas as pd
```

```python
df=pd.read_csv("temperatures.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
df.columns
```

```python
x=df['YEAR']
y=df['ANNUAL']
```

```python
plt.title("Temparature plot of India")
plt.xlabel("Year")
plt.ylabel("Annual Temperature")
plt.scatter(x,y)
```

```python
x.shape
```

```python
import numpy as np
```

```python
x=x.values
```

```python
x
```

```python
x=x.reshape(117,1)
```

```python
x.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
predicted=lr.predict(x)
```

```python
predicted
```

```python
plt.title("Temperature plot Of India")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.scatter(x,y,label='actual',color="blue")
plt.plot(x,predicted,label='predicted',color="red")
plt.legend()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y,predicted)
print("MSE=",mse)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y,predicted)
print("MAE=",mae)
```
