```python
import pandas as pd
```

```python
df=pd.read_csv("Mall_Customers.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,3:]
```

```python
x
```

```python
import matplotlib.pyplot as plt
```

```python
plt.title("Unclustered Data")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"])
```

```python
from sklearn.cluster import KMeans,AgglomerativeClustering
```

```python
SSE=[]
for i in range(1,16):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    SSE.append(km.inertia_)
```

```python
SSE
```

```python
plt.title("Elbow Method")
plt.xlabel("No. of Clusters")
plt.ylabel("Inertia Value")
plt.xticks(range(1,16))
plt.plot(range(1,16),SSE,marker=".",color="blue")
plt.grid()
```

```python
from sklearn.metrics import silhouette_score
Silh=[]
for i in range(2,16):
    km=KMeans(n_clusters=i)
    labels=km.fit_predict(x)
    score=silhouette_score(x,labels)
    Silh.append(score);
```

```python
plt.title("Silhouette Method")
plt.xlabel("No. of Clusters")
plt.ylabel("Inertia Value")
plt.xticks(range(2,16))
plt.plot(range(2,16),Silh,color="blue")
plt.grid()
```

```python
km=KMeans(n_clusters=5)
```

```python
kmean=km.fit_predict(x)
```

```python
plt.title("nclustered Data using KMeans")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"],c=kmean)
```

```python
ac=AgglomerativeClustering(n_clusters=5)
agl=ac.fit_predict(x)
```

```python
plt.title("clustered Data using AgglorometiveClustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"],c=agl)
```
